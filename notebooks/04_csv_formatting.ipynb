Fonction pour formater les CSV après conversion .fit --> CSV

Format de sortie :
Nom, Date, Latitutde, Longitude


```python
import pandas as pd
import os

FIT_file = "/Users/mgerenius/VScode_Workspace/Projet_01-Carte_dExploration/data/00_DATA_TEST/FIT--CSV-data.csv"
output_folder = "/Users/mgerenius/VScode_Workspace/Projet_01-Carte_dExploration/data/00_DATA_TEST"


def function_3(file, output_folder):
    df = pd.read_csv(file)
    file_name = os.path.basename(file)

    data_list = []
    count = 0
    for row in df.itertuples():
        if pd.notnull(row.position_lat) and pd.notnull(row.position_long):
            count += 1
            date, y = row.timestamp.split(' ',1)
            data_list.append({
                "nom": file_name + str(count),
                "date": date,
                "latitude": row.position_lat,
                "longitude": row.position_long
            })
    new_df = pd.DataFrame(data_list)

    output_file = os.path.join(output_folder, file_name)

    new_df.to_csv(output_folder + "/" + "format" + file_name, index=False, encoding="utf-8")

    print(f"le dataframe a bien été créé : {new_df}")


function_3(FIT_file, output_folder)
```

```python
import pandas as pd
import os

GPX_file = "/Users/mgerenius/VScode_Workspace/Projet_01-Carte_dExploration/data/00_DATA_TEST/GPX--CSV-data.csv"
output_folder = "/Users/mgerenius/VScode_Workspace/Projet_01-Carte_dExploration/data/00_DATA_TEST"


def function_4(file, output_folder):
    df = pd.read_csv(file)
    file_name = os.path.basename(file)

    data_list = []
    count = 0
    for row in df.itertuples():
        count += 1
        date, y = row.time.split(' ',1)
        data_list.append({
            "nom": file_name + str(count),
            "date": date,
            "latitude": row.latitude,
            "longitude": row.longitude
        })
    new_df = pd.DataFrame(data_list)

    output_file = os.path.join(output_folder, file_name)

    new_df.to_csv(output_folder + "/" + "format" + file_name, index=False, encoding="utf-8")

    print(f"le dataframe a bien été créé : {new_df}")


function_4(GPX_file, output_folder)
```
