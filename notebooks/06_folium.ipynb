```python
import os
import pandas as pd

INPUT_FOLDER = "/Users/mgerenius/VScode_Workspace/Projet_01-Carte_dExploration/data/CSV_files"

def check_files(input_folder):
    # Lister et trier les fichiers dans l'ordre alphabétique/numérique croissant
    csv_files = sorted(os.listdir(input_folder))

    empty_counter = 0
    problematic_files = []

    for csv_file in csv_files:
        csv_file_path = os.path.join(input_folder, csv_file)

        try:
            df = pd.read_csv(csv_file_path)
            if df.empty:
                print(f"Le fichier {csv_file} est vide.")
                empty_counter += 1
        except pd.errors.EmptyDataError:
            print(f"Le fichier {csv_file} ne contient pas de données exploitables.")
            problematic_files.append(csv_file)
        except UnicodeDecodeError:
            print(f"Le fichier {csv_file} a un problème d'encodage.")
            problematic_files.append(csv_file)
        except Exception as e:
            print(f"Erreur lors de la lecture du fichier {csv_file}: {e}")
            problematic_files.append(csv_file)

    print(f"Nombre de fichiers vides : {empty_counter}")
    print(f"Fichiers problématiques : {problematic_files}")

# Appel de la fonction
check_files(INPUT_FOLDER)
```

```python
import folium
import pandas as pd

def function_3(csv_file):
    map = folium.Map()

    try:
        if csv_file.endswith(".csv"):
            # Création d'un dataframe
            df = pd.read_csv(csv_file)

            # Vérifiez que les colonnes nécessaires existent
            if 'latitude' not in df.columns or 'longitude' not in df.columns:
                print("Les colonnes 'latitude' et 'longitude' sont manquantes dans le fichier CSV.")
                return map

            # Collecte des coordonnées GPS de chaque ligne du fichier CSV
            for row in df.itertuples():
                try:
                    latitude = float(row.latitude)
                    longitude = float(row.longitude)
                    # Création de la balise GPS associée
                    folium.Marker(location=[latitude, longitude]).add_to(map)
                except ValueError:
                    print(f"Coordonnées invalides : {row}")
        else:
            print(f"Le fichier {csv_file} n'est pas un fichier CSV.")

    except pd.errors.EmptyDataError:
        print(f"Le fichier {csv_file} est inexploitable.")
    except Exception as e:
        print(f"Le fichier {csv_file} a rencontré un problème lors de sa lecture: {e}")

    return map

INPUT_FILE = "/Users/mgerenius/VScode_Workspace/Projet_01-Carte_dExploration/data/CSV_files/11408573058.csv"

# Appel de la fonction
map_result = function_3(INPUT_FILE)
map_result
```

```python
import folium
from folium.plugins import HeatMap
from folium.plugins import MiniMap
import pandas as pd



def MAP(csv_file):
  """
  Génère une carte du monde en y placant des balises aux coordonnées (lat, long) présentes dans les fichiers CSV passés en revue
  """
  # Création de la carte avec Folium
  map = folium.Map(location=(48.8622893, 2.3526954), tiles='OpenStreetMap.Mapnik', zoom_start = 8)


  df = pd.read_csv(csv_file)   # Dataframe
  list_GPS = []   # Liste stockant les couples [latitude, longitude]

  # Passage en revue de chacune des lignes du dataframe
  for row in df.itertuples():

    try:
      latitude = row.latitude
      longitude = row.longitude
      list_GPS.append([latitude, longitude])

    except:
      pass    # Si le format du CSV n'est pas correct, évite une erreur

  # Création d'une 'HeatMap' à partir des coordonnées récoltées et ajout sur la carte
  HeatMap(list_GPS,
          max_opacity=.4,
          radius=9,
          blur=17
    ).add_to(map)


  # Ajout de fonctionnalités :

  # Plein écran
  folium.plugins.Fullscreen(
              position="topright",
              title="Plein écran",
              title_cancel="Réduire",
              force_separate_button=True,
    ).add_to(map)

  # Mini carte
  MiniMap().add_to(map)

  return map



### Appel de la fonction ###

CSV_FILE = '/Users/mgerenius/VScode_Workspace/Projet_01-Carte_dExploration/data/FINAL_data/FINAL_REDUCED_DATA.csv'

MAP = MAP(CSV_FILE)
MAP
```
